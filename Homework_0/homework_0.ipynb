```python
from sklearn import datasets
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cancer = datasets.load_breast_cancer()
```

```python
cancer.keys()
```

```python
features = cancer['feature_names']
print(f"Features names:\n{features}")
```

```python
featureNames = cancer.DESCR
print(f"Feature description:\n{featureNames}")
```

```python
numberOfInstances = cancer.data.shape[0]
numberOfFeatures = cancer.data.shape[1]

print(f"Number of instances: {numberOfInstances} & Number of features: {numberOfFeatures}")
```

```python
print(f"Cancer data shape: {cancer.data.shape}")
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target
df.describe()
```

```python
print(df['target'].value_counts())
df['target_name'] = df['target'].map({0: 'malignant', 1: 'benign'})
print(df['target_name'].value_counts())
```

c) Make a scatterplot of ‘mean compactness’ and ‘mean perimeter’. Try to color the points by their label (i.e. malignant or benign). You may find scatter (from matplotlib) or jointplot (from seaborn) useful for this.

```python
scatter_data = df[['mean perimeter', 'mean compactness', 'target']]
malignant = df[df['target']==0]
benign = df[df['target']==1]

plt.scatter(malignant['mean perimeter'], malignant['mean compactness'], color='red', label='malignant')
plt.scatter(benign['mean perimeter'], benign['mean compactness'], color='green', label='malignant')


plt.xlabel('Mean Perimeter')
plt.ylabel('Mean Compactness')
plt.title('Mean Perimeter vs Mean Compactness')
```
